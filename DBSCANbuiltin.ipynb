```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Create some sample 2D points (X, Y)
data = np.array([[3, 7], [4, 6], [5, 5], [6, 4], [7, 3], [6, 2], [7, 2], [8, 4], [3, 3], [2, 6], [3, 5], [2, 4]])

# Initialize DBSCAN with eps (neighborhood radius) and min_samples (min points to form a cluster)
dbscan = DBSCAN(eps=1.9, min_samples=4)

# Fit the DBSCAN model to the data
dbscan.fit(data)

# Get the cluster labels (-1 means noise/outliers)
labels = dbscan.labels_

# Get the core sample indices (indices of core points)
core_sample_indices = dbscan.core_sample_indices_

# Identify core points, border points, and noise points
core_points = data[core_sample_indices]
border_points = data[(labels != -1) & (~np.isin(np.arange(len(data)), core_sample_indices))]
noise_points = data[labels == -1]

# Plotting the core points, border points, and noise points
plt.scatter(core_points[:, 0], core_points[:, 1], color='green', marker='o', label='Core Points')
plt.scatter(border_points[:, 0], border_points[:, 1], color='blue', marker='x', label='Border Points')
plt.scatter(noise_points[:, 0], noise_points[:, 1], color='red', marker='s', label='Noise Points')

plt.title('DBSCAN Clustering with Core, Border, and Noise Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
